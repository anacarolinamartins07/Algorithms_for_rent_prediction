## Prevendo o valor do aluguel *(Machine learning supervisionado)*, com algoritmo para regressão (KNN).

```python
import pandas as pd #importando biblioteca para ler os dados
```

```python
# Criando um Data frame com a importação de um arquivo csv, para fazermos o tratamento dos dados contidos no arquivo.
df = pd.read_csv("/content/aluguel.csv", sep=";")
df.head() #serve para nos mostrar como está o df
```

```python
df.describe() #serve para nos mostrar informações como média, mínimo, máximo etc.
```

```python
#definindo minhas features
x = df[["area"]]
y = df["valor_aluguel"]
```

```python
#tecnica hold out
from sklearn.model_selection import train_test_split #importando biblioteca para treino
xtreino, xteste, ytreino, yteste= train_test_split(x,y, test_size=0.25, random_state=42) #dividindo os dados para treino e teste
```

```python
#modelagem dos dados
from sklearn.neighbors import KNeighborsRegressor #importando a biblioteca knn

k=16
knn = KNeighborsRegressor(n_neighbors = k)

knn.fit(xtreino,ytreino) #treinando o modelo

y_hat= knn.predict(xteste) #validando o modelo
```

```python
#erro quadrático médio
from sklearn.metrics import mean_squared_error # importando biblioteca do erro quadrático médio

eqm= mean_squared_error(yteste, y_hat) #calculando o erro
print('O eqm foi: ', round(eqm,2))
```

### incluindo outra váriavel (quartos)

```python
x = df[["area","quartos"]]
y = df["valor_aluguel"]
```

```python
xtreino, xteste, ytreino, yteste = train_test_split (x, y, test_size = 0.25, random_state=42) #random_state mantem o resultado mesmo que o codigo seja reproduzido várias vezes
```

```python
knn = KNeighborsRegressor(n_neighbors = k )

knn.fit(xtreino,ytreino,)

y_hat= knn.predict(xteste)
```

```python
eqm= mean_squared_error(yteste, y_hat)
print('O eqm foi: ', round(eqm,2))
```

### incluindo mais uma váriavel (mobiliaria)

```python
x = df[["area","quartos","mobiliaria"]]
y = df["valor_aluguel"]
```

```python
xtreino, xteste, ytreino, yteste = train_test_split (x, y, test_size = 0.25, random_state=42)
```

```python
knn = KNeighborsRegressor(n_neighbors = k )

knn.fit(xtreino,ytreino)

y_hat= knn.predict(xteste)
```

```python
eqm= mean_squared_error(yteste, y_hat)
print('O eqm foi: ', round(eqm,2))
```

```python
#variáveis para armazenar o melhor valor de K e o menor EQM
melhor_k = None
melhor_eqm = float('inf')

# Loop para percorrer todos os valores de K de acordo com o parametro que eu colocar
for k in range(1, 21):
  knn = KNeighborsRegressor(n_neighbors=k) # Inicializando o modelo KNN
  knn.fit(xtreino, ytreino) #Treinando o modelo
  y_hat = knn.predict(xteste) #validando o modelo
  eqm = mean_squared_error(yteste, y_hat) #calculando o erro quadrático médio
  #verificando se este é o melhor k até agora:
  if eqm < melhor_eqm:
        melhor_k = k
        melhor_eqm = eqm

# Arredondar o menor EQM para 2 casas decimais
melhor_eqm_arredondado = round(melhor_eqm, 2)
```

```python
print("Melhor valor de K:", melhor_k)
print("EQM correspondente:", melhor_eqm_arredondado)
```

#### CONCLUSÃO

A cada features que adicionamos elas podem ou não melhorar a nossa previsão, dependendo do quão relevante ela é para nossa váriavel que queremos prever. Neste caso com uma quantidade maior de features o modelo ficou melhor, como podemos ver abaixo:

Com apenas uma feature esse foi o valor da previsão do aluguel: 5652303.46

Com duas features já podemos ver melhoria, o valor da previsão foi de: 5535806.24
